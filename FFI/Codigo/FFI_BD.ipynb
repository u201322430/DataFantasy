**Borrar las columnas que no usaré en el df2.
Dejar Name en ambas para que sean el key**

Columnas a eliminar: 'HP', 'Gil', 'Exp', 'Dmg', 'Run Lev.', 'Mag. Def.', 'Family', 'Weakness',	'Resists'

```
col_a_elim = ['HP', 'Gil', 'Exp', 'Dmg', 'Run Lev.', 'Mag. Def.', 'Family', 'Weakness',	'Resists']
df2.drop(col_a_elim,axis=1,inplace=True)
```

**Inner Join**

```
dfIJ = pd.merge(df1, df2, how='inner', on='Name')
dfIJ.head()
```

**Renombrar Columnas**\
Background --> Zone\
Status --> Effect\
MDef --> MDef(%)\
Wk --> Weakness\
Res --> Resist\
Sp. Att --> SpAtk

```
dfIJ = dfIJ.rename(
    columns = {'Background' : 'Zone',
               'Status' : 'Effect',
               'MDef' : 'MDef(%)',
               'Wk' : 'Weakness',
               'Res' : 'Resist',
               'Sp. Att' : 'SpAtk',})
dfIJ.head()
```

**Dividir la columna MDef / 2**

```
dfIJ['MDef(%)'] = dfIJ['MDef(%)']/2
```

```
dfIJ.head()
```

**Ordenar el dfIJ y guardarlo en dfFinal con el siguiente orden:**\
Name,	HP,	Atk,	Def,	Magic,	MDef%,	SpAtt,	Weak,	Res,	Hits,	Acc,	Status,	Crit,	Abs,	Eva,	Gil,	Exp,	Zone,	#Int,	#Agi



> Declaro las variables



```
dfIJ_Name = dfIJ.loc[:, ['Name']]
dfIJ_HP_Atk_Def = dfIJ.loc[:, ['HP','Atk','Def']]
dfIJ_Magic = dfIJ.loc[:, ['Magic']]
dfIJ_MDef = dfIJ.loc[:, ['MDef(%)']]
dfIJ_Weakness_Resist = dfIJ.loc[:, ['Weakness','Resist']]
dfIJ_Hits_Acc_Effect_Crit_Abs_Eva = dfIJ.loc[:, ['Hits','Acc','Effect','Crit','Abs','Eva']]
dfIJ_Gil_EXP = dfIJ.loc[:, ['Gil','EXP']]
dfIJ_Zone = dfIJ.loc[:, ['Zone']]
```



> **Concat** Lo uno y se lo asigno a la nueva variable



```
dfFinal = pd.concat([dfIJ_Name, dfIJ_HP_Atk_Def, dfIJ_Magic, dfIJ_MDef, dfIJ_Weakness_Resist, dfIJ_Hits_Acc_Effect_Crit_Abs_Eva, dfIJ_Gil_EXP, dfIJ_Zone],axis=1)
```

```
#HP = 35
#FRZ = 10
#AGI = 8
#INT = 1
#RES = 15
#SUE = 8
#Probabilidad aumento:  0.75, 0.625, 0.5, 0.375, 0.32
#P_F = 0.75
#P_A = 0.625
#P_I = 0.32
#P_R = 0.5
#P_S = 0.5

HP = 25
FRZ = 3
AGI = 5
INT = 20
RES = 2
SUE = 10
P_F = 0.32
P_A = 0.32
P_I = 0.75
P_R = 0.32
P_S = 0.32
```

```
import math
import random
```

```
def prob_aum(prob):
    if prob > random.random():
        return 1
    else:
        return 0
```

```
print('Nv.\t' + 'HP\t' + 'FRZ\t' + 'AGI\t' + 'INT\t' + 'RES\t' + 'SUE')
n = 250
for i in range(1, n):   # n-1 iteraciones (1-50 = 49), n prints con el último
    print(str(i) + '\t' + str(HP) + '\t' + str(FRZ) + '\t' + str(AGI) + '\t' + str(INT) + '\t' + str(RES) + '\t' + str(SUE))
    HP = HP + math.trunc(RES/4) + 1 + prob_aum(0.5)*random.randint(20, 25)
    FRZ = FRZ + prob_aum(P_F)
    AGI = AGI + prob_aum(P_A)
    INT = INT + prob_aum(P_I)
    RES = RES + prob_aum(P_R)
    SUE = SUE + prob_aum(P_S)
print(str(n) + '\t' + str(HP) + '\t' + str(FRZ) + '\t' + str(AGI) + '\t' + str(INT) + '\t' + str(RES) + '\t' + str(SUE))
```

## Código para tabla de EXP y creación de dataframe dfXP

```
data = []
varA,varB,varC = -1,2,-2
for i in range(99):
    varC = varC + 1
    varB += varC
    varA += varB
    data.append({'Nv':i + 1, 'XP': 14 * varA})
dfXP = pd.DataFrame(data)
dfXP
```
